```python
import matplotlib.pyplot as plt
import numpy as np

# Data
indices = ['0', '1', '2']
geometry_avg_fps = [27.17, 28.96, 28.93]
geometry_low_fps = [24.34, 24.78, 25.02]
bunny_avg_fps = [34.35, 35.31, 38.84]
bunny_low_fps = [31.58, 33.35, 35.80]

# Set up the figure and axes
plt.figure(figsize=(14, 8))
ax = plt.gca()

# Bar width
width = 0.2

# Positions for bars
x = np.arange(len(indices))

# Plot bars
rects1 = ax.bar(x - 1.5*width, geometry_avg_fps, width, label='Geometry Avg FPS', color='skyblue')
rects2 = ax.bar(x - 0.5*width, geometry_low_fps, width, label='Geometry 1% Low FPS', color='lightblue')
rects3 = ax.bar(x + 0.5*width, bunny_avg_fps, width, label='Bunny Avg FPS', color='lightgreen')
rects4 = ax.bar(x + 1.5*width, bunny_low_fps, width, label='Bunny 1% Low FPS', color='palegreen')

# Customize the plot
ax.set_ylabel('Frames per second', fontsize=18)
ax.set_xlabel('Scene Index', fontsize=18)
# ax.set_title('Average FPS and 1% Low FPS Comparison', fontsize=22)
ax.set_xticks(x)
ax.set_xticklabels(indices, fontsize=18)
ax.tick_params(axis='y', labelsize=18)
ax.legend(fontsize=18)

# Set y-axis to start from 15
y_min = 15
y_max = max(max(geometry_avg_fps), max(bunny_avg_fps)) + 5
ax.set_ylim(y_min, y_max)

# Add a line at y=15 to indicate the break in the y-axis
ax.axhline(y=y_min, color='gray', linestyle='--', linewidth=1)

# Add value labels on top of each bar
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=16)

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
data_structures = ['Ground Truth', 'Octree', 'MLP', 'SIREN']
geometry_sizes = [8388608, 16661624, 270044, 799300]
bunny_sizes = [8388608, 7738808, 270044, 799300]
geometry_compression = [1.00, 0.50, 31.06, 10.49]
bunny_compression = [1.00, 1.08, 31.06, 10.49]

# Function to format numbers in scientific notation with Unicode superscripts
def format_scientific(number):
    exponent = int(np.log10(number))
    mantissa = number / 10**exponent
    # Unicode superscript digits
    superscripts = str.maketrans("0123456789", "")
    return f"{mantissa:.0f}  10{str(exponent).translate(superscripts)}"

# Function to format compression ratios
def format_compression(number):
    return f"{number:.1f}"

# Set up the figure and axes
fig, ax1 = plt.subplots(figsize=(20, 8))
ax2 = ax1.twinx()  # Create a second y-axis

# Bar width
width = 0.2

# Positions for bars
x = np.arange(len(data_structures))

# Plot storage size bars
rects1 = ax1.bar(x - width/2, geometry_sizes, width, label='Geometry Size', color='skyblue', alpha=0.7)
rects2 = ax1.bar(x + width/2, bunny_sizes, width, label='Bunny Size', color='lightgreen', alpha=0.7)

# Plot compression ratio points
scatter1 = ax2.scatter(x - width/2, geometry_compression, color='blue', s=50, label='Geometry Compression')
scatter2 = ax2.scatter(x + width/2, bunny_compression, color='green', s=50, label='Bunny Compression')

# Customize the plot
ax1.set_ylabel('Storage Size (Bytes)', fontsize=18)
ax1.set_yscale('log')
ax2.set_ylabel('Compression Ratio', fontsize=18)
ax2.set_yscale('log')  # Set compression ratio to log scale
ax1.set_xticks(x)
ax1.set_xticklabels(data_structures, fontsize=18, ha='right')

# Combine legends
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper right', fontsize=18)

# Add value labels
def autolabel(rects, ax):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(format_scientific(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=16)

autolabel(rects1, ax1)
autolabel(rects2, ax1)

for i, v in enumerate(geometry_compression):
    ax2.annotate(format_compression(v), (x[i] - width/2, v), xytext=(5, 5), 
                 textcoords='offset points', ha='left', va='bottom', fontsize=14)
for i, v in enumerate(bunny_compression):
    ax2.annotate(format_compression(v), (x[i] + width/2, v), xytext=(5, 5), 
                 textcoords='offset points', ha='left', va='bottom', fontsize=14)

# Set y-axis limits for compression ratio
ax2.set_ylim(0.1, 100)  # Adjust as needed

# Add gridlines for the compression ratio axis
ax2.grid(True, which='minor', linestyle='--', color='gray', alpha=0.3)

# Adjust layout and display
plt.tight_layout()
plt.show()
```
